```python
mainurl = "http://music.163.com/api/song/lyric?os=pc&lv=-1&kv=-1&tv=-1&id="
urls = [mainurl + id for id in ids]
for url in urls[0:3]:
    print url
```

```python
import requests

r = requests.get(urls[0])
print r.json()
```

```python
jobs = [gevent.spawn(requests.get, url) for url in urls]
print jobs[0]
```

```python
gevent.joinall(jobs)

results = [job.value for job in jobs]
print 'done'
```
